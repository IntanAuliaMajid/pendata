**Memahami Penyakit Ginjal Kronis (Chronic Kidney Disease - CKD)**

Penyakit Ginjal Kronis (PGK) atau Chronic Kidney Disease (CKD) adalah kondisi di mana ginjal mengalami kerusakan dan kehilangan kemampuannya untuk berfungsi dengan baik seiring waktu, biasanya selama berbulan-bulan atau bertahun-tahun. Ginjal yang sehat berfungsi menyaring limbah dan kelebihan cairan dari darah, yang kemudian dikeluarkan dari tubuh sebagai urine. Ketika ginjal rusak, limbah menumpuk di dalam tubuh, yang dapat menyebabkan masalah kesehatan serius.

**Apa Itu Ginjal?**

Ginjal adalah sepasang organ berbentuk kacang yang terletak di kedua sisi tulang belakang, tepat di bawah tulang rusuk. Fungsi utamanya adalah:

Menyaring darah dan membuang produk limbah dan kelebihan cairan.
Mengatur tekanan darah.
Memproduksi hormon yang diperlukan untuk produksi sel darah merah, kesehatan tulang, dan fungsi tubuh lainnya.

**Penyebab Umum CKD:**
CKD seringkali disebabkan oleh kondisi medis jangka panjang yang menekan ginjal. Beberapa penyebab paling umum meliputi:

Diabetes Mellitus (DM): Ini adalah penyebab utama CKD. Kadar gula darah tinggi yang tidak terkontrol merusak pembuluh darah di ginjal dari waktu ke waktu.
Hipertensi (HTN) / Tekanan Darah Tinggi: Tekanan darah tinggi dapat merusak pembuluh darah kecil di ginjal, mengurangi kemampuannya untuk menyaring darah.
Penyakit Jantung (termasuk Penyakit Arteri Koroner - CAD): Kondisi jantung dapat memengaruhi aliran darah ke ginjal.
Glomerulonefritis: Peradangan pada filter kecil di ginjal (glomeruli).
Penyakit Ginjal Polikistik: Kelainan genetik yang menyebabkan banyak kista (kantong berisi cairan) tumbuh di ginjal.
Obstruksi Saluran Kemih Jangka Panjang: Misalnya, dari batu ginjal atau pembesaran prostat.
Penggunaan Obat-obatan Tertentu: Beberapa obat, jika digunakan dalam jangka panjang, dapat merusak ginjal.

**Gejala CKD:**
Pada tahap awal, CKD seringkali tidak menunjukkan gejala. Gejala cenderung muncul ketika kerusakan ginjal semakin parah. Gejala dapat meliputi:

Kelelahan, kurang energi
Kesulitan tidur
Perubahan frekuensi buang air kecil
Bengkak di pergelangan kaki dan kaki (edema pedal - pe)
Otot kram
Kulit gatal atau kering
Kehilangan nafsu makan (appet)
Mual dan muntah
Sakit kepala
Mati rasa atau kesemutan
Perubahan status mental (konsentrasi)
Anemia (ane) karena ginjal tidak memproduksi cukup eritropoietin (hormon yang merangsang produksi sel darah merah).

**Keterkaitan dengan Fitur Dataset:**
Banyak fitur dalam dataset kita secara langsung berkaitan dengan indikator CKD atau kondisi penyebabnya:

age: Usia adalah faktor risiko.
bp (tekanan darah), htn (hipertensi), dm (diabetes mellitus), cad (penyakit arteri koroner): Ini adalah penyebab utama CKD yang ada dalam dataset.
sg (berat jenis), al (albumin), su (gula), rbc (sel darah merah), pc (sel nanah), pcc (gumpalan sel nanah), ba (bakteri): Ini adalah parameter urin yang mengindikasikan kerusakan ginjal atau infeksi saluran kemih.
bgr (glukosa darah acak), bu (urea darah), sc (kreatinin serum), sod (natrium), pot (kalium): Ini adalah parameter darah yang menunjukkan fungsi ginjal dan keseimbangan elektrolit. Kadar urea dan kreatinin yang tinggi adalah indikator utama fungsi ginjal yang buruk.
hemo (hemoglobin), pcv (packed cell volume), wbcc (jumlah sel darah putih), rbcc (jumlah sel darah merah): Ini adalah parameter darah lengkap, di mana anemia (rendahnya hemoglobin dan sel darah merah) adalah komplikasi umum CKD. Jumlah sel darah putih dapat mengindikasikan infeksi.
appet (nafsu makan), pe (edema pedal), ane (anemia): Ini adalah gejala klinis CKD yang juga ada dalam dataset.

### Dataset ini berisi berbagai fitur klinis dan demografi pasien yang dapat digunakan untuk memprediksi atau memahami Penyakit Ginjal Kronis (CKD).

id  
Tipe Data: Numerik (Integer)  
Deskripsi Singkat: Pengidentifikasi unik untuk setiap catatan pasien. Kolom ini biasanya tidak digunakan dalam analisis prediktif karena tidak membawa informasi medis, tetapi penting untuk mengidentifikasi baris data.
Potensi Masalah Data: Duplikasi ID (jarang, tetapi mungkin terjadi).  

age (usia)  
Tipe Data: Numerik (Integer)  
Deskripsi Singkat: Usia pasien dalam tahun.  
Potensi Masalah Data: Nilai yang hilang, nilai di luar rentang usia yang wajar.  

bp (blood pressure - tekanan darah)  
Tipe Data: Numerik (Integer)  
Deskripsi Singkat: Tekanan darah diastolik atau sistolik pasien (tergantung pada konvensi dataset, umumnya tekanan sistolik atau rata-rata).  
Potensi Masalah Data: Nilai yang hilang, nilai yang tidak realistis (misalnya, tekanan darah sangat rendah atau sangat tinggi).  

sg (specific gravity - berat jenis urin)  
Tipe Data: Numerik (Float)  
Deskripsi Singkat: Berat jenis urin, yang mengukur konsentrasi partikel dalam urin. Nilai yang lebih rendah dapat mengindikasikan ketidakmampuan ginjal untuk mengkonsentrasikan urin.  
Nilai yang Mungkin: Contoh: 1.005, 1.010, 1.015, 1.020, 1.025.  
Potensi Masalah Data: Nilai yang hilang, nilai yang tidak berada dalam rentang normal (sekitar 1.003-1.030).  

al (albumin - albumin dalam urin)  
Tipe Data: Kategorikal (Integer atau Ordinal)  
Deskripsi Singkat: Tingkat albumin dalam urin. Albumin dalam urin (proteinuria) adalah indikator penting kerusakan ginjal.  
Nilai yang Mungkin: 0 (tidak ada), 1, 2, 3, 4, 5 (menunjukkan tingkat keparahan yang meningkat).  
Potensi Masalah Data: Nilai yang hilang, nilai di luar rentang yang diharapkan (0-5).  

su (sugar - gula dalam urin)  
Tipe Data: Kategorikal (Integer atau Ordinal)  
Deskripsi Singkat: Tingkat gula dalam urin (glikosuria). Kehadiran gula dalam urin dapat mengindikasikan diabetes, yang merupakan faktor risiko utama CKD.  
Nilai yang Mungkin: 0 (tidak ada), 1, 2, 3, 4, 5 (menunjukkan tingkat keparahan yang meningkat).  
Potensi Masalah Data: Nilai yang hilang, nilai di luar rentang yang diharapkan (0-5).  

rbc (red blood cells - sel darah merah dalam urin)  
Tipe Data: Kategorikal (String/Object)  
Deskripsi Singkat: Kehadiran sel darah merah dalam urin (hematuria).  
Nilai yang Mungkin: 'normal', 'abnormal'.  
Potensi Masalah Data: Nilai yang hilang, salah ketik ('norml', 'abnorml'), format yang tidak konsisten.  

pc (pus cell - sel nanah/leukosit dalam urin)  
Tipe Data: Kategorikal (String/Object)  
Deskripsi Singkat: Kehadiran sel nanah/leukosit dalam urin (piuria).  
Nilai yang Mungkin: 'normal', 'abnormal'.  
Potensi Masalah Data: Nilai yang hilang, salah ketik, format yang tidak konsisten.  

pcc (pus cell clumps - gumpalan sel nanah)  
Tipe Data: Kategorikal (String/Object)  
Deskripsi Singkat: Kehadiran gumpalan sel nanah dalam urin.  
Nilai yang Mungkin: 'present', 'notpresent'.  
Potensi Masalah Data: Nilai yang hilang, salah ketik, format yang tidak konsisten.  

ba (bacteria - bakteri dalam urin)  
Tipe Data: Kategorikal (String/Object)  
Deskripsi Singkat: Kehadiran bakteri dalam urin (bakteriuria).  
Nilai yang Mungkin: 'present', 'notpresent'.  
Potensi Masalah Data: Nilai yang hilang, salah ketik, format yang tidak konsisten.  

bgr (blood glucose random - gula darah acak)  
Tipe Data: Numerik (Integer)  
Deskripsi Singkat: Tingkat gula darah acak pasien. Indikator diabetes.  
Potensi Masalah Data: Nilai yang hilang, nilai yang tidak realistis (misalnya, sangat rendah atau sangat tinggi).  

bu (blood urea - urea darah)  
Tipe Data: Numerik (Float)  
Deskripsi Singkat: Tingkat urea dalam darah. Urea adalah produk limbah yang disaring oleh ginjal. Tingkat tinggi dapat menunjukkan fungsi ginjal yang buruk.  
Potensi Masalah Data: Nilai yang hilang, nilai yang tidak realistis.  

sc (serum creatinine - kreatinin serum)  
Tipe Data: Numerik (Float)  
Deskripsi Singkat: Tingkat kreatinin dalam serum darah. Kreatinin adalah produk limbah otot yang disaring oleh ginjal.  
Tingkat tinggi adalah indikator penting gangguan fungsi ginjal.  
Potensi Masalah Data: Nilai yang hilang, nilai yang tidak realistis.  

sod (sodium - natrium)  
Tipe Data: Numerik (Float)  
Deskripsi Singkat: Tingkat natrium dalam serum darah. Ketidakseimbangan natrium dapat terjadi pada penyakit ginjal.  
Potensi Masalah Data: Nilai yang hilang, nilai yang tidak realistis.  

pot (potassium - kalium)  
Tipe Data: Numerik (Float)  
Deskripsi Singkat: Tingkat kalium dalam serum darah. Tingkat kalium yang tidak normal dapat berbahaya dan sering terkait dengan masalah ginjal.  
Potensi Masalah Data: Nilai yang hilang, nilai yang tidak realistis.  

hemo (hemoglobin - hemoglobin)  
Tipe Data: Numerik (Float)  
Deskripsi Singkat: Tingkat hemoglobin dalam darah. Anemia (hemoglobin rendah) sering terjadi pada pasien CKD.  
Potensi Masalah Data: Nilai yang hilang, nilai yang tidak realistis.  

pcv (packed cell volume - volume sel darah merah)  
Tipe Data: Numerik (Integer)  
Deskripsi Singkat: Volume sel darah merah dalam darah, juga indikator anemia.  
Potensi Masalah Data: Nilai yang hilang, nilai yang tidak realistis, mungkin dalam format string jika ada nilai non-numerik.  

wc (white blood cell count - jumlah sel darah putih)  
Tipe Data: Numerik (Integer)  
Deskripsi Singkat: Jumlah sel darah putih. Peningkatan dapat mengindikasikan infeksi.  
Potensi Masalah Data: Nilai yang hilang, nilai yang tidak realistis, mungkin dalam format string jika ada nilai non-numerik (misalnya, '6200' vs. 6200).  

rc (red blood cell count - jumlah sel darah merah)  
Tipe Data: Numerik (Float)  
Deskripsi Singkat: Jumlah sel darah merah. Rendah dapat mengindikasikan anemia.  
Potensi Masalah Data: Nilai yang hilang, nilai yang tidak realistis, mungkin dalam format string jika ada nilai non-numerik.  

htn (hypertension - hipertensi/tekanan darah tinggi)  
Tipe Data: Kategorikal (String/Object)  
Deskripsi Singkat: Indikator apakah pasien memiliki riwayat hipertensi. Hipertensi adalah faktor risiko utama dan komplikasi CKD.  
Nilai yang Mungkin: 'yes', 'no'.  
Potensi Masalah Data: Nilai yang hilang, salah ketik, format yang tidak konsisten.  

dm (diabetes mellitus - diabetes melitus)  
Tipe Data: Kategorikal (String/Object)  
Deskripsi Singkat: Indikator apakah pasien memiliki riwayat diabetes melitus. Diabetes adalah penyebab umum CKD.  
Nilai yang Mungkin: 'yes', 'no'.  
Potensi Masalah Data: Nilai yang hilang, salah ketik, format yang tidak konsisten (misalnya, ' yes', 'no ').  

cad (coronary artery disease - penyakit arteri koroner)  
Tipe Data: Kategorikal (String/Object)  
Deskripsi Singkat: Indikator apakah pasien memiliki riwayat penyakit arteri koroner. Penyakit kardiovaskular sering menyertai CKD.  
Nilai yang Mungkin: 'yes', 'no'.  
Potensi Masalah Data: Nilai yang hilang, salah ketik, format yang tidak konsisten.  

appet (appetite - nafsu makan)  
Tipe Data: Kategorikal (String/Object)  
Deskripsi Singkat: Kondisi nafsu makan pasien. Penurunan nafsu makan adalah gejala umum pada pasien CKD yang parah.
Nilai yang Mungkin: 'good', 'poor'.  
Potensi Masalah Data: Nilai yang hilang, salah ketik, format yang tidak konsisten.  

pe (pedal edema - edema perifer/pembengkakan kaki)  
Tipe Data: Kategorikal (String/Object)  
Deskripsi Singkat: Kehadiran edema perifer (pembengkakan pada kaki, pergelangan kaki, atau tungkai). Gejala umum CKD.
Nilai yang Mungkin: 'yes', 'no'.  
Potensi Masalah Data: Nilai yang hilang, salah ketik, format yang tidak konsisten.  

ane (anemia - anemia)  
Tipe Data: Kategorikal (String/Object)  
Deskripsi Singkat: Kehadiran anemia pada pasien. Anemia adalah komplikasi umum CKD.  
Nilai yang Mungkin: 'yes', 'no'.  
Potensi Masalah Data: Nilai yang hilang, salah ketik, format yang tidak konsisten.  
class (class - kelas/target)

Tipe Data: Kategorikal (String/Object)  
Deskripsi Singkat: Kolom target yang menunjukkan apakah pasien memiliki Penyakit Ginjal Kronis atau tidak.  
Nilai yang Mungkin: 'ckd' (Chronic Kidney Disease), 'notckd'.  
Potensi Masalah Data: Nilai yang hilang, salah ketik (misalnya, ' ckd', 'notckd\t'), ketidakseimbangan kelas.

```python
import pandas as pd # Untuk bekerja dengan DataFrame
import numpy as np  # Untuk operasi numerik, terutama NaN
import matplotlib.pyplot as plt # Untuk membuat plot dan grafik (ini yang disingkat 'plt')
import seaborn as sns # Untuk visualisasi data yang lebih estetis (ini yang disingkat 'sns')

# --- 1. Pemuatan Data ---
# Memuat dataset dari file CSV
try:
    df = pd.read_csv('chronic_kidney_disease.csv')
    print("Dataset berhasil dimuat.")
except FileNotFoundError:
    print("Error: 'chronic_kidney_disease.csv' tidak ditemukan. Pastikan file berada di direktori yang benar.")
    # Jika file tidak ditemukan, Anda bisa menambahkan logika untuk memuat dari UCI repo jika perlu
    # atau langsung keluar dari program.
    exit()

# --- 2. Pembersihan Data Awal ---
# Mengganti semua karakter '?' dengan nilai NaN (Not a Number)
df.replace('?', np.nan, inplace=True)

# Mengidentifikasi kolom yang seharusnya numerik tetapi mungkin masih bertipe 'object'
numerical_cols_to_convert = [
    'age', 'bp', 'sg', 'al', 'su', 'bgr', 'bu', 'sc', 'sod',
    'pot', 'hemo', 'pcv', 'wbcc', 'rbcc'
]

# Mengubah kolom-kolom tersebut menjadi tipe numerik.
# 'errors="coerce"' akan mengubah nilai yang tidak bisa dikonversi menjadi NaN.
for col in numerical_cols_to_convert:
    if col in df.columns: # Memastikan kolom ada sebelum diakses
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Membersihkan spasi di awal/akhir dan mengembalikan string 'nan' menjadi np.nan
# pada semua kolom bertipe 'object' (kategorikal)
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].astype(str).str.strip().replace('nan', np.nan)

# Khusus untuk variabel target 'class', bersihkan typo '\tckd' menjadi 'ckd'
df['class'] = df['class'].replace({'ckd\t': 'ckd'})
df['class'] = df['class'].str.strip() # Pastikan tidak ada spasi sisa

print("\nData berhasil dimuat dan dibersihkan secara awal.")
print("\nInfo DataFrame setelah pembersihan awal:")
df.info()

# --- 3. Visualisasi Nilai Hilang ---
print("\n--- Visualisasi Nilai Hilang ---")

# Visualisasi Heatmap Nilai Hilang
plt.figure(figsize=(15, 7))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Visualisasi Nilai Hilang di Dataset')
plt.ylabel('Baris Data')
plt.xlabel('Kolom Fitur')
plt.show() # Menampilkan plot

# Bar Chart Persentase Nilai Hilang per Kolom
missing_percentage = (df.isnull().sum() / len(df)) * 100
missing_percentage = missing_percentage[missing_percentage > 0].sort_values(ascending=False)

if not missing_percentage.empty:
    plt.figure(figsize=(12, 6))
    sns.barplot(x=missing_percentage.index, y=missing_percentage.values, palette='plasma')
    plt.xticks(rotation=90) # Memutar label sumbu x agar tidak tumpang tindih
    plt.title('Persentase Nilai Hilang per Kolom')
    plt.ylabel('Persentase Nilai Hilang (%)')
    plt.xlabel('Kolom')
    plt.tight_layout() # Menyesuaikan tata letak agar label tidak terpotong
    plt.show() # Menampilkan plot
else:
    print("\nTidak ada nilai hilang yang tersisa setelah pembersihan.")
```
